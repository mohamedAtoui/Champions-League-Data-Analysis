```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv(r"C:\Users\i7 11Th\Desktop\Python2Proj\final\distribution_team.csv")
#rename the col of data frame so it's more easy for the reader
data1 = data1.rename(columns={'0':'club','1':'gk_performance','2':'def_performance','3':'mid_performance','4':'att_performance'})
#round the values of all the column in the data frame
data1['gk_performance'] = data1['gk_performance'].round(2)
data1['def_performance'] = data1['def_performance'].round(2)
data1['mid_performance'] = data1['mid_performance'].round(2)
data1['att_performance'] = data1['att_performance'].round(2)
#sort the data frame by alphabitic order of teams names:
data1= data1.sort_values(by='club', ascending=True)
data1=data1.reset_index(drop=True)
print((data1))
```

```python
data_def=pd.read_csv(r"C:\Users\i7 11Th\Desktop\Python2Proj\final\avg_score_defending.csv")
#rename the col to suitable name
data_def=data_def.rename(columns={'Score':'performance_def'})
#round the values
data_def['performance_def']=data_def['performance_def'].round(2)
#add a col to data1
data1['performance_def']=data_def['performance_def']

data1
```

```python
data_gk=pd.read_csv(r"C:\Users\i7 11Th\Desktop\Python2Proj\final\avg_score_goalkeeping.csv")
#rename the col to suitable name
data_gk['Scorekeeping']=data_gk['Scorekeeping'].round(2)
data1['Scorekeeping']=data_gk['Scorekeeping']
data1
```

```python
data_on_target_per_game=pd.read_csv(r"C:\Users\i7 11Th\Desktop\Python2Proj\final\avg_on_target_result.csv")
#drop unnecessary column
data_on_target_per_game=data_on_target_per_game.drop('Unnamed: 0',axis=1)
#sort the name of the clubs in the alphabitic oreder to match it with the main table 
data_on_target_per_game= data_on_target_per_game.sort_values(by='0', ascending=True)
#rename the col to suitable names
data_on_target_per_game = data_on_target_per_game.rename(columns={
    '0':'club',
    '1':'on_target_per_match'})
#round the values of second column
data_on_target_per_game['on_target_per_match']=data_on_target_per_game['on_target_per_match'].round(2)

data1['on_target_per_match']=data_on_target_per_game['on_target_per_match']
data1
```

```python
data_assist_dribble=pd.read_csv(r"C:\Users\i7 11Th\Desktop\Python2Proj\final\assists_dribble_game.csv")

data_assist_dribble=data_assist_dribble.rename(columns={'WeightedAvg':'Assist_dribble_per_match'})

data_assist_dribble['Assist_dribble_per_match']=data_assist_dribble['Assist_dribble_per_match'].round(2)

data1['Assist_dribble_per_match']=data_assist_dribble['Assist_dribble_per_match']
data_assist_dribble
```

```python
data1['col_Sum']=data1['gk_performance']+data1['def_performance']+data1['mid_performance']+data1['att_performance']+data1['performance_def']+data1['Scorekeeping']+data1['on_target_per_match']+data1['Assist_dribble_per_match']
data=data1.copy()
data= data.sort_values(by='col_Sum', ascending=False)
data
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
pos_bars = ax.bar(data['club'], data['col_Sum'], color='teal', label='Attacking Efficiency')

ax.set_xlabel('Teams')
ax.set_ylabel('Sum')
ax.set_xticks(range(len(data['col_Sum'])))  
ax.set_xticklabels(data['club'], rotation=90) 
ax.set_title('sum of all features')
ax.axhline(0, color='grey', linewidth=0.8)  # Adds a line at y=0
ax.legend()
ax.set_yticklabels([])
# Customize Y-axis tick labels to show positive values even for negative bars

# Display the plot
plt.show()
```

```python
#scaling out of 100% on_target_per_match

#finding Max of on_target_per_match column
Max=max(data['on_target_per_match'])

#findig Min of on_target_per_match column
Min=min(data['on_target_per_match'])

#calculated scaled_ontarget using next formula:
data['scaled_on_target']=(data['on_target_per_match']-Min)/(Max-Min)
data['scaled_on_target']=data['scaled_on_target']*100

#rounded to 2 digit after
data['scaled_on_target']=data['scaled_on_target'].round(2)
#data['scaled_on_target']

#scaling out of 100% Assists_dribble_per_match

#finding Max of on_target_per_match column
Max=max(data['Assist_dribble_per_match'])

#findig Min of on_target_per_match column
Min=min(data['Assist_dribble_per_match'])

#calculated scaled_Assist_dribble_per_match using next formula:
data['scaled_Assist_dribble_per_match']=(data['Assist_dribble_per_match']-Min)/(Max-Min)
data['scaled_Assist_dribble_per_match']=data['scaled_Assist_dribble_per_match']*100

#rounded to 2 digit after
data['scaled_Assist_dribble_per_match']=data['scaled_Assist_dribble_per_match'].round(2)
#data['scaled_Assist_dribble_per_match']

data
```

```python
data['Overall']=0.05*data['gk_performance']+0.1*data['Scorekeeping']+0.05*data['def_performance']+0.1*data['performance_def']+  0.15*data['att_performance']+0.15*data['mid_performance']+0.2*data['scaled_on_target']+0.2*data['scaled_Assist_dribble_per_match']
data['Sum']
```

```python
#scaling out of 100% performance_def

#finding Max of on_target_per_match column
Max=max(data['performance_def'])

#findig Min of on_target_per_match column
Min=min(data['performance_def'])

#calculated scaled_ontarget using next formula:
data['scaled_performance_def']=(data['performance_def']-Min)/(Max-Min)
data['scaled_performance_def']=data['scaled_performance_def']*100

#rounded to 2 digit after
data['scaled_performance_def']=data['scaled_performance_def'].round(2)
#data['scaled_on_target']

#scaling out of 100% Assists_dribble_per_match

#finding Max of on_target_per_match column
Max=max(data['Scorekeeping'])

#findig Min of on_target_per_match column
Min=min(data['Scorekeeping'])

#calculated scaled_Assist_dribble_per_match using next formula:
data['scaled_Scorekeeping']=(data['Scorekeeping']-Min)/(Max-Min)
data['scaled_Scorekeeping']=data['scaled_Scorekeeping']*100

#rounded to 2 digit after
data['scaled_Scorekeeping']=data['scaled_Scorekeeping'].round(2)
#data['scaled_Assist_dribble_per_match']

data
```

```python
data['new_Overall']=0.05*data['gk_performance']+0.1*data['scaled_Scorekeeping']+0.05*data['def_performance']+0.1*data['scaled_performance_def']+  0.15*data['att_performance']+0.15*data['mid_performance']+0.2*data['scaled_on_target']+0.2*data['scaled_Assist_dribble_per_match']
data= data.sort_values(by='new_Overall', ascending=False)
data['new_Overall']=data['new_Overall'].round(2)
data
```

```python




import matplotlib.pyplot as plt

fig, ax = plt.subplots()
pos_bars = ax.bar(data['club'], data['new_Overall'], color='teal', label='Attacking Efficiency')

ax.set_xlabel('Teams')
ax.set_ylabel('Overal Score')
ax.set_xticks(range(len(data['club']))) 
ax.set_xticklabels(data['club'], rotation=90) 
ax.set_title('Comparative Analysis of Teams'' Average Pass and Cross Accuracy for Attackers and Defenders')
ax.axhline(0, color='grey', linewidth=0.8)  # Adds a line at y=0
ax.legend()
ax.set_yticklabels([])

plt.show()
```
