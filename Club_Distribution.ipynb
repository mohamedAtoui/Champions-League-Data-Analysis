```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\i7 11Th\Desktop\Python2Proj\Data\distributon.csv")
data=np.array(data)
data
```

```python
#devid data into mid/def/forward/goal_keeper. 
def_dis = [i for i in data if i[3] == 'Defender']
def_dis=np.array(def_dis)

mid_dis = [i for i in data if i[3] == 'Midfielder']
mid_dis=np.array(mid_dis)

for_dis = [i for i in data if i[3] == 'Forward']
for_dis=np.array(for_dis)

gol_dis = [i for i in data if i[3] == 'Goalkeeper']
gol_dis=np.array(gol_dis)

#the next line is for checking if there's any missing values
#so because t is eaqule to total of dataset everything is fine
t=len(def_dis)+len(mid_dis)+len(for_dis)+len(gol_dis)
#t==len(data). TRUE
def_dis
```

```python
import statistics as st

#club list
club_list=[]
for i in data[:,2]:
    if(i not in club_list):
        club_list.append(i)
club_list=np.array(club_list)

club_avrg=[[],[],[]]
club_avrg=np.array(club_avrg)

final_data=[]
#the next for loop put each player in his team by first taking goal keppers/defenders/midfielders/attackers of the team then put it together
#in 4 column array
for i in club_list:
    golk=[]
    deff=[]
    mid=[]
    forr=[]
    for j in def_dis:
        if(j[2]==i):
            deff.append(j[-1])
    for j in for_dis:
        if(j[2]==i):
            forr.append(j[-1])
    for j in mid_dis:
        if(j[2]==i):
            mid.append(j[-1])
    for j in gol_dis:
        if(j[2]==i):
            golk.append(j[-1])
    deff=st.mean(deff)
    golk=st.mean(golk)
    mid=st.mean(mid)
    forr=st.mean(forr)
    final_data.append([i,golk,deff,mid,forr])

#sort the data by attackers score
final_data=np.array(final_data)
att=final_data[:,-1]
sorted_indices = np.argsort(att)
final_data0 = final_data[sorted_indices].copy()


#sort the data by defenders score
deff=final_data[:,2]
sorted_indices = np.argsort(deff)
final_data1 = final_data[sorted_indices].copy()
print(final_data0)
final_data1
#df present the next:
#[team][mean of goal keepers][mean of defenders][mean of midfielders][mean of attackers]

            
```

```python
#[team][mean of goal keepers][mean of defenders][mean of midfielders][mean of attackers]
import matplotlib.pyplot as plt


#final_data=final_data[:-1]
teams=final_data0[:,0]
defenders=final_data0[:,2]
att=final_data0[:,-1]

#defenders=np.array(defenders)
def_vis=[]
for i in defenders:
    def_vis.append(-1*(int(float(i))))
    
att_vis=[]
for i in att:
    att_vis.append((int(float(i))))
print(len(att_vis),len(att),len(final_data))

fig, ax = plt.subplots()
pos_bars = ax.bar(teams, att_vis, color='teal', label='Attacking Efficiency')
neg_bars = ax.bar(teams, def_vis, color='red', label='Defending Efficiency')
ax.set_xlabel('Teams')
ax.set_ylabel('Values')
ax.set_xticks(range(len(teams)))  
ax.set_xticklabels(teams, rotation=90) 
ax.set_title('Comparative Analysis of Teams'' Average Pass and Cross Accuracy for Attackers and Defenders')
ax.axhline(0, color='grey', linewidth=0.8)  # Adds a line at y=0
ax.legend()
ax.set_yticklabels([])
# Customize Y-axis tick labels to show positive values even for negative bars

# Display the plot
plt.show()
```
